Step 1: Data Exploration & Leading

```python
#importing dependencies

import pandas as pd
#mysql toolkit
import pymysql #this work as adapter
from sqlalchemy import create_engine
```

```python
df=pd.read_csv('Walmart.csv',encoding_errors='ignore')
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
#All duplicated items are counted
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
#dropping all the rows with missing records of quantity and unit price
df.dropna(inplace=True)
```

```python
df.dtypes
```

```python
df['unit_price'].astype(float)#here we got error coz there is $ sign present thats why we cant so we have to deal with that first
```

```python
df['unit_price']=df['unit_price'].str.replace('$','').astype(float)
df.head()
```

```python
#here we make a new coloum of Total which will tell us total amount
df['total']=df['unit_price']*df['quantity']
df.head()
```

```python
df.info()
df.to_csv('walmart_clean_data.csv',index=False)
df.columns=df.columns.str.lower()
```

```python
#mysql connection
engine_mysql=create_engine("mysql+pymysql://root:1234@127.0.0.1:3306/Walmart_db")
try:
    engine_mysql
    print("Connected succesfuly")
except:
    print("no Connection made")
```

```python
df.to_sql(name='walmart',con=engine_mysql,if_exists='append',index=False)
```
